```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
dfs = []

path = Path("submissions")
submissions = [
    'lightgbm-finetuned.csv',
    'catb-fe-cv5.csv',
    'xgb-cv5-finetuned.csv'
]

for file in submissions:
    df = pd.read_csv(path / file)
    df = df.iloc[:, -1]            
    df.name = Path(file).stem           
    dfs.append(df)

result = pd.concat(dfs, axis=1)
```

```python
result
```

```python
import numpy as np

weights = np.array([0.2, 0.1, 0.7])  # веса для каждого столбца

new_preds = np.average(result, axis=1, weights=weights)
```

```python
submission = pd.read_csv('sample_submission.csv')
submission['Churn'] = new_preds
submission.to_csv('submission.csv', index=False)
```
